## eの出現理由
### 関数f(x) = A^xをx=0において微粉した傾きが1になるAってあるのかな
<img src='John_Napier.jpg'>


$$
f(x) = A^x
\\
f'(0) = 1
$$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 3, 100)
y = np.e**x

y2 = x + 1

plt.figure(figsize=(15, 10))
plt.rc('font')

plt.plot(x, y, label='y=A^x')
plt.plot(x, y2, label='y=x + 1')

p = plt.plot([0, 0],[-1, 20], "black") 
p = plt.plot([-1, 3],[0, 0], "black") 
plt.ylim(-1, 5)
plt.yticks(np.arange(0, 6, step=1))
plt.grid()
plt.legend(fontsize=20)
```

## 対数を使って考えてみる

$$

### これらの関数はy=xにおいて線対称になる

```python
x1 = np.linspace(-1, 3, 100)
y1 =np.e**x
y1_1 = x + 1

x2 = np.linspace(0.1, 3, 100)
y2 = np.log(x2)
y2_1 =  x1 - 1

plt.figure(figsize=(15, 10))
plt.rc('font')

plt.plot(x1, y1, label='y=A^x')
plt.plot(x1, y1_1, label='y=x+1')
plt.plot(x2, y2, label='y=logA(x)')
plt.plot(x1, y2_1,label='y=x-1')
plt.plot(x, x, label='y=x')

p = plt.plot([0, 0],[-2, 20], "black") 
p = plt.plot([-1, 3],[0, 0], "black") 
plt.ylim(-2, 5)
plt.yticks(np.arange(0, 6, step=1))
plt.grid()
plt.legend(fontsize=20)
```

### つまりy=logA(x)のx=1における接線を求めればいいのでは


### 微分の定義に基づいた微分
$$
f'(x) = \lim_{h\to0} \frac {f(x+h) -f(x)}{h}
$$

### y=logA(x)でやる
$$
f(x) = log_A(x)
\\
f'(x) = \lim_{h\to0} \frac {log_A(x+h) -log_A(x)}{h}
$$
### 1/hを前に出して、logの引き算は割り算になるので変形
$$
f'(x) = \lim_{h\to0} \frac {1}{h} log_A\frac{x+h} {x}
$$
### x=1を代入
$$
f'(1) = \lim_{h\to0} \frac {1}{h} log_A\frac{1+h} {1}
\\
f'(1) = \lim_{h\to0} \frac {1}{h} log_A(1+h)
$$
### logの前にあるものは指数にできるので変形したものが1になる
$$
f'(1) = \lim_{h\to0}  log_A(1+h)^\frac {1}{h} = 1
$$

## logが1になるためには底(A)と値が同じになる必要がある
### 例
$$
2 = 2^1 <=>  {lox_2}2 = 1
$$

### どっかで見たぞこの式
$$
A = \lim_{h\to0}  (1+h)^\frac {1}{h}
$$
